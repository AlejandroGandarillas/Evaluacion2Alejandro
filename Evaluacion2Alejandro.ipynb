```python
#Parte 1
import numpy as np


array = np.random.randint(1, 101, 10)
print("Array generado:", array)


matriz = array.reshape(2, 5)
print("\nMatriz 2x5:")
print(matriz)


primera_fila = matriz[0, :]
ultima_columna = matriz[:, -1]
print("\nPrimera fila:", primera_fila)
print("\nÚltima columna:", ultima_columna)


matriz[1, :] = 0
print("\nMatriz:")
print(matriz)
```

```python
#Parte 2 


array = np.random.randint(1, 101, 10)
print("Array generado:", array)


matriz = array.reshape(2, 5)
print("\nMatriz 2x5:")
print(matriz)


primera_fila = matriz[0, :]
ultima_columna = matriz[:, -1]
print("\nPrimera fila:", primera_fila)
print("\nÚltima columna:", ultima_columna)


matriz[1, :] = 0
print("\nMatriz modificada:")
print(matriz)


array1 = np.random.randint(1, 51, 5)
array2 = np.random.randint(1, 51, 5)
print("\nArray 1:", array1)
print("Array 2:", array2)

suma_arrays = array1 + array2
producto_arrays = array1 * array2
print("\nSuma de los arrays:", suma_arrays)
print("Producto de los arrays:", producto_arrays)

promedio = np.mean(suma_arrays)
desviacion_estandar = np.std(suma_arrays)
print("\nPromedio de la suma de los arrays:", promedio)
print("Desviación estándar de la suma de los arrays:", desviacion_estandar)


array_random = np.random.rand(20)
print("\nArray de números aleatorios entre 0 y 1:", array_random)

array_random[array_random < 0.5] = 0
array_random[array_random >= 0.5] = 1
print("\nArray modificado con condiciones lógicas:", array_random)


array3 = np.random.randint(1, 11, (3, 3))
array4 = np.random.randint(1, 11, (3, 3))
print("\nArray 3:")
print(array3)
print("\nArray 4:")
print(array4)

producto_escalar = np.dot(array3, array4)
print("\nProducto escalar de los arrays 3 y 4:")
print(producto_escalar)
```

```python
#Parte 3
matriz_datos = np.random.randint(1, 21, (5, 4))
print("\nMatriz de datos (5 filas, 4 columnas):")
print(matriz_datos)


media_columnas = np.mean(matriz_datos, axis=0)
print("\nMedia de cada columna:", media_columnas)

valor_maximo_filas = np.max(matriz_datos, axis=1)
valor_minimo_filas = np.min(matriz_datos, axis=1)
print("\nValor máximo de cada fila:", valor_maximo_filas)
print("Valor mínimo de cada fila:", valor_minimo_filas)

matriz_ordenada = matriz_datos[matriz_datos[:, 1].argsort()]
print("\nMatriz ordenada según los valores de la segunda columna (peso):")
print(matriz_ordenada)
```
